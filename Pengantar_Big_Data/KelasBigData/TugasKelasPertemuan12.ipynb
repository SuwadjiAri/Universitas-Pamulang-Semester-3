```python
import urllib
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
with urllib.request.urlopen("https://jalantikus.com/tips/aplikasi-chatting/") as page:
    html = page.read()
    print(html)
```

```python
soup = bs(html, 'lxml')
print(type(soup))
print(soup.prettify())
```

```python
texts = soup.find_all("p")
for item in texts:
    desc = item.get_text();
    print(desc)    
```

```python
list_rem = []
list_desc = []
texts = soup.find_all("p")
for i, item in enumerate(texts):
    desc = item.get_text();
    if (i == 0 or i == 6 or i == 9 or i == 12 or i > 28):
#         print (i, desc);
#         print('tidak perlu dimasukkan ke list')
        list_rem.append(desc)
    else :
        list_desc.append(desc)
    
print(list_desc)
```

```python
# Proses Lemmatisasi
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
wnl = WordNetLemmatizer()
list1 = list_desc[0].split(" ")
# print(list1)
for words in list1:
    print(words + " ---> " + wnl.lemmatize(words))
```

```python
# Proses Stemming
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
ps = PorterStemmer()
words = list1
for x in words:
    print(x, " : ", ps.stem(x))
```
