```python
import re
import string
frequency = dict()
document_text = open('Hasil Final Komentar.csv')
text_string = document_text.read()
match_pattern = re.findall(r'\b[a-z]{3,15}\b', text_string)
for word in match_pattern:
    count = frequency.get(word, 0)
    frequency[word] = count + 1
    
frequency_list = frequency.keys()
for word in frequency_list :
    print(word)
    print (frequency[word])
    
```

```python
import pandas as pd
import numpy as np
kata = []
jumlah_kata = []

for word in match_pattern :
    count = frequency.get(word, 0)
    frequency[word] = count
    
    kata.append(word)
    jumlah_kata.append(frequency[word])
    
data = {'kata': kata, 'count': jumlah_kata}
df = pd.DataFrame(data, columns = ['kata', 'count'])
```
