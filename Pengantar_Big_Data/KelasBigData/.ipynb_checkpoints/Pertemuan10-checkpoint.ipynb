```python
import urllib
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
keyword = 'mouse'
with urllib.request.urlopen("https://bhinneka.com/jual?cari="+keyword) as page:
    html = page.read()
    print(html)
```

```python
soup = bs(html, 'lxml')
print(type(soup))
print(soup.prettify())
```

```python
produk = soup.find_all("div", "o_wsale_product_grid_wrapper position-relative h-100")
i = 1
for i in produk :
    nama = i.find('h6', "o_wsale_products_item_title mb-2").get_text().replace('\n','')
    harga = i.find('span', 'h6 mb-0').get_text().replace('Rp','').replace(' ', '').replace('.','')
    print(nama)
    print(harga)
```

```python
nama_produk = []
harga_produk = []

produk = soup.find_all("div", "o_wsale_product_grid_wrapper position-relative h-100")
i = 1
for i in produk :
    nama = i.find('h6', "o_wsale_products_item_title mb-2").get_text().replace('\n','')
    harga = i.find('span', 'h6 mb-0').get_text().replace('Rp','').replace(' ', '').replace('.','')
    nama_produk.append(nama)
    harga_produk.append(int(harga))
    
data = {'nama': nama_produk, 'harga': harga_produk}
df = pd.DataFrame(data, columns = ['nama', 'harga'])

# df
```

```python
df.to_csv("DataScrapping.csv", sep=',')
```
