```python
import urllib
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
keyword='pulsar'
with urllib.request.urlopen("https://www.bukalapak.com/products?from=omnisearch&from_keyword_history=false&search%5Bkeywords%5D="+keyword+"&search_source=omnisearch_keyword&source=navbar") as page:
    html = page.read()
    print(html)
```

```python
soup = bs(html, 'lxml')
print(type(soup))
print(soup.prettify())
```

```python
produk = soup.find_all("div", "bl-product-card__description")
# produk
i = 1
for i in produk :
    nama = i.find('a', "bl-link").get_text().replace('\n','').replace('  ','')
    harga = i.find('p', 'bl-text bl-text--subheading-20 bl-text--semi-bold bl-text--ellipsis__1').get_text().replace('Rp','').replace(' ', '').replace('.','').replace('\n','')
    print(nama)
    print(harga)
```

```python
nama_produk = []
harga_produk = []
produk = soup.find_all("div", "bl-product-card__description")
# produk
i = 1
for i in produk :
    nama = i.find('a', "bl-link").get_text().replace('\n','').replace('  ','')
    harga = i.find('p', 'bl-text bl-text--subheading-20 bl-text--semi-bold bl-text--ellipsis__1').get_text().replace('Rp','').replace(' ', '').replace('.','').replace('\n','')
#     print(nama)
#     print(harga)
    nama_produk.append(nama)
    harga_produk.append(int(harga))
    
data = {'nama': nama_produk, 'harga': harga_produk}
df = pd.DataFrame(data, columns = ['nama', 'harga'])

df
```

```python
df.to_csv("DataScrapping_Bukalapak_pulsar.csv", sep=',')
```
